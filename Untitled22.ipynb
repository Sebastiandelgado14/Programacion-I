En Python existen distintos elementos para guardar y organizar información, y cada uno se usa según la necesidad.

Variables
Se usan cuando queremos guardar un solo valor (un número, un texto, un resultado).
Sirven para reutilizar datos y hacer el código más claro.

Listas
Se usan cuando necesitamos guardar varios valores ordenados y que puedan cambiar.
Son ideales cuando los datos pueden crecer, achicarse o modificarse.

Tuplas
Se usan cuando queremos guardar varios valores ordenados, pero que no deban modificarse.
Aportan seguridad: los datos no cambian accidentalmente.

Diccionarios
Se usan cuando necesitamos relacionar una clave con un valor.
Son útiles cuando los datos tienen significado (por ejemplo: nombre, edad, nota).

Conjuntos (sets)
Se usan cuando queremos guardar valores sin repetir y sin importar el orden.
Son útiles para comparar grupos y eliminar duplicados.

## **CONTROL DE FLUJO EN PYTHON**

```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a = 0
b = 1
while b < 100:
  print(b)
  a, b  = b, a+b # El valor de b no está actualizado
```

* La primer línea contiene una asignación múltiple: las variables a y b toman en forma simultanea los nuevos valores 0 y 1. En la última linea esto se vuelve a usar, demostrando que las expresiones a la derecha son evaluadas antes de que suceda cualquier asignación. Las expresiones a la derecha son evaluadas de izquierda a derecha.
* El bucle while se ejecuta mientras la condición (aquí: b < 100) sea verdadera. La prueba usada en el ejemplo es una comparación simple. Los operadores estándar de comparación se
escriben:
< (menor qué),
> (mayor qué),
== (igual a),
<= (menor o igual qué),
>= (mayor o igual qué)
!= (distinto a).
* El cuerpo del bucle está sangrado: la sangría es la forma que usa Python para agrupar declaraciones. En el intérprete interactivo se debe teclear un tab o espacio(s) para cada línea sangrada. En la práctica, un buen editor, tiene la facilidad de agregar la sangría automáticamente. Al ingresar una declaración compuesta en forma interactiva, debe finalizar con una línea en blanco para indicar que está completa (ya que el analizador no puede adivinar cuando tecleaste la última línea).

**Nota:** a, b  = b, a+b en una sola línea, hace que el valor de b sea tomado dos pasos antes, que es lo que busca la secuencia de Fibonacci. Sin embargo, si se quiere actualizar b con el paso anterior, se toma b = a+b en otra línea.

```python
a = 0
b = 1
while b < 100:
  print(b)
  a = b
  b = a+b # El valor de b está actualizado
```

En caso de que queramos que se realicen máximo 3 intentos, podemos realizar el siguiente programa:

```python
clave_correcta = "python123"
intentos = 3
while intentos > 0:
    intento = input("Ingresa la contraseña: ")
    if intento == clave_correcta:
        print("¡Acceso concedido!")
        break
    else:
        intentos -= 1
        print("Intentos restantes:", intentos)
else:
    print("Cuenta bloqueada")
```

```python
while True:
    print("MENÚ")
    print("1. Saludar")
    print("2. Sumar dos números")
    print("3. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        print("¡Hola!")

    elif opcion == "2":
        a = int(input("Número 1: "))
        b = int(input("Número 2: "))
        print("Resultado:", a + b)

    elif opcion == "3":
        print("Adiós")
        break

    else:
        print("Opción inválida")
```

```python
try:
    numero = int(input("Ingresa un número: "))
    resultado = 10 / numero
    print("Resultado:", resultado)
except ZeroDivisionError:
    print("No se puede dividir entre cero")
except ValueError:
    print("Debes ingresar un número válido")
finally:
    print("Fin del programa")
```

```python
numero = int(input("Ingresa un número: "))

if numero < 2:
    print("No es primo")
else:
    es_primo = True

    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            es_primo = False
            break

    if es_primo:
        print("Es primo")
    else:
        print("No es primo")
```

```python
print("ADIVINA EL NÚMERO")
print("(Entre 1 y 5, escribe 0 para salir)")

numero_secreto = 3

while True:
    pass
    intento = int(input("Adivina: "))
    if intento == 0:
        print("¡Hasta luego!")
        break
    if intento < 1 or intento > 5:
        print("Número fuera de rango, intenta de nuevo")
        continue
    if intento == numero_secreto:
        print("¡Correcto! Adivinaste")
        break
    else:
        print("Incorrecto, sigue intentando")
```
