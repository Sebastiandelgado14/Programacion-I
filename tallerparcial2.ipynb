```python
import numpy as np
import pandas as pd

np.random.seed(123) # para guardar la semilla
```

```python
datos = np.random.normal(loc = 4.0, scale = 0.5, size = 50)

df = pd.DataFrame({
    "nota": datos
})
```

```python
df. describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde

# Datos
data = df["nota"].dropna()

# Parámetros de la normal
mu = np.mean(data)
sigma = np.std(data)

# Rango para graficar
x = np.linspace(min(data), max(data), 1000)

# FDP normal
pdf_normal = norm.pdf(x, mu, sigma)

# KDE (Kernel Density Estimation)
kde = gaussian_kde(data)
pdf_kde = kde(x)

# Gráfica
plt.figure()

# Histograma (densidad)
plt.hist(data, bins=15, density=True)

# Curva normal
plt.plot(x, pdf_normal, label="Normal")

# Curva KDE
plt.plot(x, pdf_kde, linestyle='--', label="KDE")

# Etiquetas
plt.title("Histograma + FDP Normal + KDE")
plt.xlabel("Nota")
plt.ylabel("Densidad")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde

# Datos
data = df["nota"].dropna()

# Parámetros
mu = np.mean(data)
sigma = np.std(data)

# Rango
x = np.linspace(min(data), max(data), 1000)

# FDP normal
pdf_normal = norm.pdf(x, mu, sigma)

# KDE
kde = gaussian_kde(data)
pdf_kde = kde(x)

# Estilo general
plt.figure(figsize=(10, 6))

# Histograma más suave
plt.hist(data, bins='auto', density=True, alpha=0.6, edgecolor='black')

# Curvas
plt.plot(x, pdf_normal, linewidth=2, label="Distribución Normal")
plt.plot(x, pdf_kde, linestyle='--', linewidth=2, label="KDE (Kernel)")

# Líneas de referencia
plt.axvline(mu, linestyle=':', linewidth=2, label=f"Media = {mu:.2f}")

# Detalles
plt.title("Distribución de Notas", fontsize=16)
plt.xlabel("Nota", fontsize=12)
plt.ylabel("Densidad", fontsize=12)

plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```
